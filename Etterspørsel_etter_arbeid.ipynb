```python
import sympy as sp
import numpy as np 
from matplotlib import pyplot as plt
```

```python
def prod(L,K,a):
    return L**a*K

x = np.linspace(0.001,100,1000)

fig, ax = plt.subplots()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.plot(x,prod(x,10,.5), label = 'Produksjon')
ax.legend(loc = 'best');
ax.set_title('Produksjonstilbud');

L,K,a = sp.symbols('L K a')

gp = sp.diff(prod(L,K,a),L)

print('grenseproduktviteten til K er postiv:')
display(gp)
```
