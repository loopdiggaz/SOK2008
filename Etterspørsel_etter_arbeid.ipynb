```python
import sympy as sp
import numpy as np 
from matplotlib import pyplot as plt
```

```python
def prod(L,K):
    return L*K

x = np.linspace(0.001,100,1000)

fig, ax = plt.subplots()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.plot(x,prod(10,x), label = 'Produksjon')
ax.legend(loc = 'best');
ax.set_title('Produksjonstilbud');

L,K = sp.symbols('L K')

gp = sp.diff(prod(L,K),L)

print('grenseproduktviteten til K er postiv:')
gp
```
