```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```


Subsidisert barnepass:

Subsidert barnepass vil føre til at færre kvinner velger å gå ut av arbeidsmarkdet (lavere faste kostnader).
Om substitusjonseffekten domninerer vil subsidierte variable kostnader føre til at kvinner med barn jobber flere timer.
Når et individ får barn, endres preferanser for konsum og fritid.
Alternativkostnaden for å jobbe øker under en tidsperiode, reservasjonslønna øker: W går opp.
jo eldre barne blir desto lavere blir alternativ kostnaden for å være i jobb, reservasjonslønna synker tilbake.

```python
# definerer symboler jeg kan få bruk for. 
C,L,a,b,u,m,M,P,W,T,V,u_0 = sp.symbols("C L a b u m M P W T V u_0")

L = np.linspace(0.001,5,1000)

def budsjett(L):
    return 50-10*L

def create_ax():
    fig, ax = plt.subplots()
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.set_xlabel('Antall barn')
    ax.set_ylabel('Konsum')
    ax.set_xlim(0,5)
    ax.set_ylim(0,50)

    return fig, ax

fig, ax = create_ax()

ax.plot(L,budsjett(L))
```

```python
u = C**a*L**b
L_ind_sol= sp.solve(u-u_0,L)
indiff_L=sp.lambdify((u_0,a,b,C), L_ind_sol)
indiff_L(u_0,a,b,C)
```

```python
ax.plot(L_0,budsjett(L_0))

ax.vlines(84,0,100, color='black',ls='dashed')
ax.hlines(100,0,84, color='black',ls='dashed')
ax.annotate('Arbeidsfri inntekt = m', (0+5,100+5))

u = C**a*L**b

# Løser med hensyn på L 
L_ind_sol=sp.solve(u-u_0,L)[0]
L_ind_sol

# Må gjøre om for å plotte 
indiff_L=sp.lambdify((u_0,a,b,C), L_ind_sol)
indiff_L(u_0,a,b,C)

for i in range (91, 111,6):
    ax.plot(L_0, indiff_L(i,0.5,0.5,L_0))
```
