```python
def etterspørsel(L):
    return 250-0.5*L
    
def tilbud(L):
    return 10 + 0.5*L

import numpy as np
from matplotlib import pyplot as plt
import sympy as sp

L = np.linspace(0.001,1000,10000)


def create_ax():
    fig, ax = plt.subplots()
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.set_xlim(0,1000)
    ax.set_ylim(0,500)
    
    return fig, ax

fig, ax = create_ax()

ax.plot(L,etterspørsel(L))
ax.plot(L,tilbud(L))
ax.hlines(150,0,200,linestyles='dashed')
ax.vlines(200,0,150,linestyles='dashed');

minstelønn = 150

L_minstelønn = (250-150)/0.5
L_minstelønn
```
